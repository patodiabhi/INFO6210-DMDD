```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

soup = BeautifulSoup

i=1

#Lists to store scraped value data

airplane_name = []
airplane_details = []

#while loop for getting data from all the pages of url (Pagination)

while i<6:

    page = requests.get("https://bestcompany.com/airlines?page="+str(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    i=i+1    
    airplanes = soup.findAll(class_='company-card card list-complete-item')
    container = airplanes[0]

    for container in airplanes:
        
        name = container.find(class_='d-md-none vc-company-name').getText()
        airplane_name.append(name)

        details = container.find(class_='d-md-none key-details').getText()
        airplane_details.append(details)
        
        #Creating dataframe and storing the scrapped data
        
        pd.set_option('display.max_colwidth', -1)
        Baggage_Details = pd.DataFrame({"Airplane Name":airplane_name,"Airplane Details":airplane_details})

Baggage_Details
```

```python
# Scrapping data using Twitter Api

import twitter

#Defininig all the required keys

CONSUMER_KEY = '########################'
CONSUMER_SECRET = '###########################'
OAUTH_TOKEN = '##########################'
OAUTH_TOKEN_SECRET = '#########################'

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,
                          CONSUMER_KEY, CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)
print(twitter_api)

#List of companies from scrapped data

flight_lists = ["SouthwestAir","JetBlue","Delta","HawaiianAir","AlaskaAir", "AmericanAir",
                "Allegiant","United","VirginAmerica","FlyFrontier", "SpiritAirlines","EndeavorAir","AmericanAir", 
                "flysaa", "cathaypacific","British_Airways", "emirates", "lufthansa","AirCanada",
                "airfrance","qatarairways","SunCountryAir","TurkishAirlines","FlyANA_official","EtihadAirways",
                "EVAAirUS","KLM","SingaporeAir","WestJet","FlyAirNZ","AirAsia","AsianaAirlines","JALFlightInfo_e",
                "FlySWISS","_austrian","HainanAirlines","MAS","SkyWestAirlines","ThaiAirways","AirEuropa",
                "bangkok_airways","IndonesiaGaruda","GulfAir","dragonair","expressjet","Qantas","airchina",
                "azalofficial","KoreanAir_KE","Fly_Norwegian"]

#List to store scraped value data in:

airplane_name = []
twitter_id = []
followers = []
friends = []
verify = []



for i in flight_lists:
    twitter_id.append(i)
    airplane_name.append(twitter_api.statuses.user_timeline(screen_name=i)[1]['user']['name'])
    followers.append(twitter_api.statuses.user_timeline(screen_name = i)[1]['user']['followers_count'])
    friends.append(twitter_api.statuses.user_timeline(screen_name = i)[1]['user']['friends_count'])
    verify.append(twitter_api.statuses.user_timeline(screen_name = i)[1]['user']['verified'])

#Creating dataframe and storing the scrapped data
    
Twitter_Specifications = pd.DataFrame({
             'Airplane Name' : airplane_name,
             'Twitter ID' : twitter_id,
             'Followers Count' : followers,
             'Friends Count' : friends,
             'Verified' : verify})

Twitter_Specifications
```
