# Getting Airplane Tweets 


```python

airplane_name = []
user_list = []
status_list = []
tweetid_list = []
createdat_list = []
fav = []


for i in flight_lists:
    results= twitter_api.statuses.user_timeline(screen_name = i)
    for status in results:
        user_list.append(i)
        airplane_name.append(status['user']['name'])
        status_list.append(status['text'])
        tweetid_list.append(status['id'])
        createdat_list.append(status['created_at'])
        fav.append(status['favorite_count'])
                   
pd.set_option('display.max_colwidth', -1)
df = pd.DataFrame(
    
   {
    'Airplane_Name':airplane_name,
    'Twitter_ID': user_list,
    'Tweet_ID' : tweetid_list,
    'Status': status_list,
    'Fav_Count':fav,
    'Time': createdat_list,
    
   })
df
```

```python

import numpy as np
import yelp
import requests
import pandas as pd

from pandas.io.json import json_normalize


API_KEY = "U23S8oPF91hwt8yO4QE4xUU9h4DED0ISEZGV8Lp3FR0kkfNWMo-3xCajXYrFaDyqhI8mqOSxeEJCuHhZnStoL2jtZlHVETKQWbKyyqiDDK8Zg3EdruFethbZWCpzXHYx"

API_HOST = 'https://api.yelp.com'
BUSINESS_PATH = '/v3/businesses/'

def get_business(business_id):
    business_path = BUSINESS_PATH + business_id
    url = API_HOST + business_path + '/reviews'
    headers = {'Authorization': f"Bearer {API_KEY}"}

    response = requests.get(url, headers=headers)

    return response.json()



flight_lists_name = ["Southwest Airlines","JetBlue Airways","Delta","Hawaiian Airlines","Alaska Airlines", "American Airlines",
                "Allegiant","United Airlines","Virgin America","Frontier Airlines"]

flight_lists = ["southwest-airlines-los-angeles-5","jetblue-airways-new-york","delta-air-lines-college-park-2","hawaiian-airlines-honolulu","alaska-airlines-seattle-10","american-airlines-los-angeles-5","allegiant-air-honolulu","united-airlines-san-francisco-6","virgin-america-san-francisco-5","frontier-airlines-denver-4"]


review_dict = []
user_name_dict = []
rating_dict = []
timecreated_dict = []

review1_dict1 = []
user1_name_dict1 = []
rating1_dict1 = []
timecreated1_dict1 = []
   
total_dict = []    
        
for i in flight_lists:
    
    json_data = get_business(i)
    results = json_data['reviews']
    
    total = json_data['total']
    total_dict.append(total)
    
#Review 1   

    name = results[0]
    review = name['text']
    review_dict.append(review)
    
    rating = name['rating']
    rating_dict.append(rating)
    
    timecreated = name['time_created']
    timecreated_dict.append(timecreated)
    
    userdet = name['user']
    username= userdet['name']
    user_name_dict.append(username)

#Review 2    

    name1 = results[1]
    review1 = name1['text']
    review1_dict1.append(review1)
    
    rating1 = name1['rating']
    rating1_dict1.append(rating1)
    
    timecreated1 = name1['time_created']
    timecreated1_dict1.append(timecreated1)
    
    userdet1 = name1['user']
    username1= userdet1['name']
    user1_name_dict1.append(username1)
    
    
     
dtf = pd.DataFrame({
              'Airplane_Name' : flight_lists_name,
              'Review_One' : review_dict,
              'Rating' : rating_dict,
              'Time_Created' : timecreated_dict,
              'User_Name' : user_name_dict,
              'Review_Two' : review1_dict1,
              'Rating2' : rating1_dict1,
              'Time_Created2' : timecreated1_dict1,
              'User_Name2' : user1_name_dict1,
              'Total_Number_Review':total_dict  })             

dtf
```
